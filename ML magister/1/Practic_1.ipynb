# Pandas


https://dfedorov.spb.ru/pandas/cheatsheet/Pandas_DataFrame_Notes.pdf


```python
import pandas as pd

# Series
salaries = pd.Series([400, 300, 200, 250], index=["Andrew", "Bob", "Charles", "Ann"])
print(salaries)
```

```python
salaries[salaries > 250]
```

```python
print(salaries.Andrew == salaries["Andrew"])
```

```python
print("Second element of salaries is", salaries[1], "\n")
# Smart indexing
print(salaries[:3], "\n")
print("There are", len(salaries[salaries > 0]), "positive elements in salaries\n")
# Series obects can be the arguments for Numpy functions
print(np.exp(salaries))
```

```python
df = pd.read_csv("./data/bank.csv", sep=",")
df
```

```python
df.describe()
```

```python
df.sort_values(by="balance", ascending=False).head()
```

```python
df[(df["balance"] > df["balance"].median()) & (df["duration"]> 100)].head()
```

```python
df.loc[(df["balance"] > df["balance"].median()) & (df["duration"]> 100),['contact','education']].head()
```

```python
g = df.groupby("education")
for (i, sub_df) in g:
    print(sub_df["balance"].mean(), sub_df["pdays"].mean())
```

```python
g.mean()
```

```python
pd.plotting.scatter_matrix(
    df[["balance", "age", "duration", "pdays"]], figsize=(15, 15), diagonal="kde")
```
